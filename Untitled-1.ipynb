```python
import matplotlib.pyplot as plt
import networkx as nx
import random


def generar_grafo():
    grafo = nx.Graph()
    for i in range(100):
        grafo.add_node(i)
    
    matriz_adyacencia = [[0 for _ in range(100)]for _ in range(100)]
   
    for nodo in grafo:
        num_amigos = random.randint(5, 10)
    
        while len(list(grafo.neighbors(nodo))) < num_amigos:
            amigo = random.randint(0, 99)

            if amigo != nodo and not grafo.has_edge(nodo, amigo):
                grafo.add_edge(nodo, amigo)
                matriz_adyacencia[nodo][amigo] = 1
                matriz_adyacencia[amigo][nodo] = 1
    
    return grafo, matriz_adyacencia

    

def amigos_en_comun(nodo1, nodo2, matriz_adyacencia):
    amigos_comun = []
    for i in range(100):
        if matriz_adyacencia[nodo1][i] == 1 and matriz_adyacencia[nodo2][i] == 1:
            amigos_comun.append(i)

    return amigos_comun

def guardar_conexiones(nodo1, nodo2, amigos_comun, grafo):
    conexiones = []
    for i in range(len(amigos_comun)):
        conexiones.append((nodo1, amigos_comun[i]))
        conexiones.append((nodo2, amigos_comun[i]))
    return conexiones




Grafo = generar_grafo()
persona1 = 4
persona2 = 5

amigos_en_comun = amigos_en_comun(persona1, persona2, Grafo[1])

print(f"la persona {persona1} y la persona {persona2} son ambos amigos de la o las personas: {amigos_en_comun}")

conexiones = guardar_conexiones(persona1, persona2, amigos_en_comun, Grafo[1])

pos = nx.spring_layout(Grafo[0])
nx.draw(Grafo[0], pos, with_labels=True)
nx.draw_networkx_edges(Grafo[0], pos, edgelist=conexiones, edge_color="red")
plt.show
```
